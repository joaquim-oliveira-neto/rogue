```python
import requests
import pandas as pd 
import numpy as np
```

```python

def api_push_keywords(tmdb_id):
    api_key = 'c71955d50f611a31406968f5d6a45beb'
    try:
        response = requests.get(f'https://api.themoviedb.org/3/movie/{tmdb_id}/keywords?api_key={api_key}')
        if response.status_code != 200:
            return None 
        
        response = response.json()
        x=0
        keywords = []
        while x < len(response['keywords']):
            for i in response['keywords']:
                keywords.append(i['name'])
                x+=1
        print(tmdb_id)        
        return keywords
    except:
        return None
```

```python
links = pd.read_csv('../raw_data/ml-25m/links.csv').dropna()
```

```python
tmdb_keywords = links.copy()
```

```python
tmdb_keywords['tmdbId'] = tmdb_keywords['tmdbId'].astype(np.int32)
```

```python
tmdb_keywords['keywords'] = tmdb_keywords['tmdbId'].apply(api_push_keywords)
```

```python
tmdb_keywords.to_csv('../raw_data/tmdb_api_keywords.csv')
```
